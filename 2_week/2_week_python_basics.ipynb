```python
""" 
You are provided with a dataset generated by an imperfect crawler that downloaded the

WeST webpage for two levels of hierarchy. It consists of:

The order of the links accessed
The crawled URLs
The amount of time in seconds to extract further URLs from that URL
The hierarchy of the link
Write a simple python script that does the following:

Read the dataset from dataset.csv file as a pandas Dataframe.
Plot the time distribution.
Remove the last 20% from your dataset.
Plot the time distribution again
Remove all the entries that has "0" in the time_in_seconds column.
Plot the time distribution again.
NOTE: For this question, you need to use pandas and may use Matplotlib only. No other library usage is permissible. 

Submit your solution in form of Python Notebook
"""
```

```python
import pandas as pd
df = pd.read_csv("./dataset.csv", index_col=0)
```

```python
df.head()
```

```python
df.info()
```

```python

import matplotlib.pyplot as plt
# Plot the initial time distribution
plt.figure(figsize=(20, 10))
plt.hist(df['time_in_seconds'], bins=60, edgecolor='black')
plt.title("Initial Time Distribution")
plt.xlabel("Time in Seconds")
plt.ylabel("Frequency")
plt.show()
```

```python
df[df["time_in_seconds"]>60]
```

```python
# Remove the last 20% from your dataset.

df_last_filtered_index = int(len(df)*0.8)
df_filtered = df.iloc[:df_last_filtered_index]
```

```python
# Plot the time distribution again
plt.figure(figsize=(20, 10))
plt.hist(df_filtered['time_in_seconds'], bins=60, edgecolor='black')
plt.title("Initial Time Distribution")
plt.xlabel("Time in Seconds")
plt.ylabel("Frequency")
plt.show()


# last 20% of the data contributed to time_in_seconds >20 seconds
# have some url's with time = 0
```

```python
len(df_filtered)
```

```python
# Remove all the entries that has "0" in the time_in_seconds column.
# Plot the time distribution again.

df_without_zeros_time_seconds = df_filtered[df_filtered["time_in_seconds"]!=0]
len(df_without_zeros_time_seconds)
```

```python
# Plot the time distribution again
plt.figure(figsize=(20, 10))
plt.hist(df_without_zeros_time_seconds['time_in_seconds'], bins=60, edgecolor='black')
plt.title("Initial Time Distribution")
plt.xlabel("Time in Seconds")
plt.ylabel("Frequency")
plt.show()
```
